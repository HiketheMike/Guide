### <span Style='Color:blue'> 2. Load Extra Functions

```python
# Write a function to clean the Missing values
def clean_data(df):
    categorical_columns = df.select_dtypes(exclude = 'number').columns
    numeric_columns = df.select_dtypes(include = 'number').columns
    
    df[categorical_columns] = df[categorical_columns].fillna('None')
    df[numeric_columns] = df[numeric_columns].fillna(0)
    return df
```

### <span Style='Color:blue'>  1. Clean the data and prepare the training data

```python
# Clean the data
df_cleaned = clean_data(df)
test_cleaned = clean_data(test)

# Define the predictors and response
X = df_cleaned.drop(columns = ['Id', 'SalePrice'])
y = df_cleaned['SalePrice']
final_features = test_cleaned.drop(columns = ['Id'])

# Separate the training and the testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    random_state = 42,
                                                    test_size = 0.2)
```

```python
# Choose the column
numeric_column = df.select_dtypes(include = 'number').drop(columns= ['Id', 'SalePrice']).columns
categorical_column = df.select_dtypes(exclude = 'number').columns

# Scale the data
scaler = StandardScaler()
X_train[numeric_column] = scaler.fit_transform(X_train[numeric_column])
X_test[numeric_column] = scaler.transform(X_test[numeric_column])
final_features[numeric_column] = scaler.transform(final_features[numeric_column])

# Encode the data
encoders = {}
for col in categorical_column:
    oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value = -1)
    X_train[col] = oe.fit_transform(X_train[[col]])
    X_test[col] = oe.transform(X_test[[col]])
    final_features[col] = oe.transform(final_features[[col]])
# Turn to tensors
train_features = torch.tensor(X_train.values, dtype = torch.float32)
train_target = torch.tensor(y_train.values, dtype = torch.float32)
test_features = torch.tensor(X_test.values, dtype = torch.float32)
final_features = torch.tensor(final_features.values, dtype = torch.float32)
```

```python
# Define the architecture of a neural network
class HousePriceRegression(nn.Module):
    def __init__(self, input_size):
        super(HousePriceRegression, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Flatten(),
            nn.Linear(input_size, 256),
            nn.ReLU())
        self.layer2 = nn.Sequential(
            nn.Linear(256, 128),
            nn.Dropout(0),
            nn.ReLU())
        self._forward = nn.Sequential(
            self.layer1,
            self.layer2,
            nn.Linear(128, 1))
    def forward(self, x):
        return self._forward(x)

# Instantiate the Neural Network
house_nn = HousePriceRegression(input_size = X_train.shape[1])
```

```python
summary(house_nn, 
        input_data = train_features,
        col_names= ['input_size',
                    'output_size',
                    'num_params'])
```

```python
# Define the criterion and optimizer
optimizer = torch.optim.Adam(params = house_nn.parameters(), lr = 0.075)
criterion = nn.MSELoss()

# Define the training loop
epochs = 2000
for epoch in range(epochs):
    house_nn.train()
    optimizer.zero_grad()
    output = house_nn(train_features).squeeze()
    loss = criterion(train_target, output)
    loss.backward()
    optimizer.step()
    if epoch % (epochs // 10) == 0:
        print(f'Epoch {epoch + 1}/{epochs}, Loss {loss:.0f}')
```

```python
# Make the evaluation 
with torch.no_grad():
    predicted = house_nn(test_features)
    rmse = root_mean_squared_error(y_test, predicted)
    print(f'Test MSE: {rmse:.3f}') 
```

### <span Style='Color:blue'> 5. Predict test data

```python
y_predicted = house_nn(final_features)
y_predicted = y_predicted.detach().numpy()
y_predicted
```

```python
# Make the submission data
submit_df = pd.DataFrame(y_predicted, index=test['Id'])
submit_df = submit_df.reset_index()
submit_df = submit_df.rename(columns={0: 'SalePrice'})
submit_df.to_csv('submit_df.csv')
```
